```python
import pandas as pd
import numpy as np
from faker import Faker
import random


num_patients=len(df)
# Set the number of patients
# num_patients = 10
unique_numbers = random.sample(range(100000, 100000+num_patients), num_patients)

# Add the prefix "P" to each number
patient_ids = ['P' + str(num) for num in unique_numbers]

# Display the list of numbers
# print(numbers_with_prefix)


# Generate patient IDs
# patient_ids = np.arange(1, num_patients+1)


patient_names =[]
for i in range(num_patients):
    patient_names.append(fake.name())
    
# Define the list of drugs
drugs = ['Metformin', 'Insulin', 'Sulfonylureas', 'GLP-1 Agonists', 'DPP-4 Inhibitors',
         'Glucocorticoids', 'Statins', 'Antipsychotics', 'Thiazide diuretics',
         'Immunosuppressants', 'Beta-blockers']

# Generate random medication data
max_medication_count = 3  # Maximum number of medications per patient
medication_counts = np.random.randint(1, max_medication_count + 1, size=num_patients)
medication_data = []
for count in medication_counts:
    medications = np.random.choice(drugs, size=count, replace=False)
    periods = np.random.randint(1, 24, size=count)  # Random period of usage (1-12 months)
    medication_data.append(list(zip(medications, periods)))

# Create the DataFrame
data = {
    'Patient_ID': [],
    'Medication': [],
    'Period_of_Usage': []
}

for patient_id, meds in zip(patient_ids, medication_data):
    data['Patient_ID'].extend([patient_id] * len(meds))
    data['Patient_Name'].extend([])
    data['Medication'].extend([med[0] for med in meds])
    data['Period_of_Usage'].extend([med[1] for med in meds])

df = pd.DataFrame(data)
df.head()
```
